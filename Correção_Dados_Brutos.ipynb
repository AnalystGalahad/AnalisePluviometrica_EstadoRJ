```python
import pandas as pd
from datetime import datetime, timedelta
import shutil
from os import remove, mkdir
```

```python
#Importação das tabelas:
ano_bissexto = ['2000', '2004', '2008', '2012', '2016', '2020']
anos = ['1997','1998','1999','2000','2001','2002','2003','2004','2005','2006',
        '2007','2008','2009','2010','2011','2012','2013','2014','2015','2016',
        '2017','2018','2019','2020','2021','2022']

#Entrada:
colunas = ['Dia', 'Hora','HBV','15 min','01 h','04 h','24 h','96 h']
pasta = input('Digite a Estação: ')

#Listas finais
Fev31_15min = []
FevNorm_15min = []

Fev31N_15min = []
FevNormN_15min = []

#Diretórios:
diretorio = r'D:\Estudo\Oficina de Pesquisa\teste'
diretorio_tratado = r'D:/Estudo/Oficina de Pesquisa/Dados Tradados'


for ano in anos:
    print(ano)
    #Tratando as entradas:
    tamanho = pasta.split()
    if len(tamanho) != 1:
        arquivo = ('_'.join(tamanho)).lower()
    else:
        arquivo = pasta.lower()
    
    #arquivo ='av_brasil_mendanha'
        
    
    mes_fev = pd.read_fwf(diretorio+'\\'+arquivo+'_'+ano+'02_Plv.txt',
                          colspecs = [(0,10),(12,20),(22,25),(28,34),(35,41),(42,48),(49,55),(56,62)],
                          infer_datetime_format=True, enconding='iso-8850-1',
                          names = colunas)
    mes_mar = pd.read_fwf(diretorio+'\\'+arquivo+'_'+ano+'03_Plv.txt',
                          colspecs = [(0,10),(12,20),(22,25),(28,34),(35,41),(42,48),(49,55),(56,62)],
                          infer_datetime_format=True, enconding='iso-8850-1',
                          names = colunas)
    
    #Concatenando as Tabelas:
    mes_concat = pd.concat([mes_fev, mes_mar], ignore_index = False)

    
    #Condição:
    if ano in ano_bissexto:
        print('É bissexto:')
        #meses[['Dia']] = meses[['Dia']].apply(pd.to_datetime)
        mes_concat['Dia'] = pd.to_datetime(mes_concat['Dia'], dayfirst= True, infer_datetime_format=True)
        meses_id = mes_concat.set_index('Dia')
        fev31 = meses_id.loc[: ano+"-03-02"]
        fevNormal = meses_id.loc[: ano+"-02-29"]
    else:
        print('Não é bissexto')
        #meses[['Dia']] = meses[['Dia']].apply(pd.to_datetime)
        mes_concat['Dia'] = pd.to_datetime(mes_concat['Dia'], dayfirst= True, infer_datetime_format=True)
        meses_id = mes_concat.set_index('Dia')
        fev31 = meses_id.loc[: ano+"-03-03"]
        fevNormal = meses_id.loc[: ano+"-02-28"]
    
    fev31 = fev31.reset_index()
    fevNormal = fevNormal.reset_index()
    
    #Fevi31
    fev31[fev31['15 min'] == 'ND'] = 0 
    fev31[['15 min']] = fev31[['15 min']].apply(pd.to_numeric)
    soma = fev31['15 min'].sum()
    norm = soma/31
    Fev31_15min.append(soma)
    Fev31N_15min.append(norm)
    

    #FevNormal
    fevNormal[fevNormal['15 min'] == 'ND'] = 0 
    fevNormal[['15 min']] = fevNormal[['15 min']].apply(pd.to_numeric)
    soma = fevNormal['15 min'].sum()
    FevNorm_15min.append(soma)
    if ano in ano_bissexto:
        norm =  soma/29
        FevNormN_15min.append(norm)
    else:
        norm = soma/28
        FevNormN_15min.append(norm)
        
    #Exportando DataFrame em csv:
    lista_de_tuplas = list(zip(anos,Fev31_15min,FevNorm_15min,Fev31N_15min, FevNormN_15min))
    df = pd.DataFrame(lista_de_tuplas, columns=['Anos','Fev31','FevNatural','Fev31N','FevNaturalN'])
    #df.to_csv(diretorio+'\\'+arquivo+'_FinalN.csv', index = False)
df
```

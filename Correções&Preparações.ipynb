# Importações:

```python
import pandas as pd
from datetime import datetime, timedelta
import shutil
#from os import remove, mkdir
from tqdm import tqdm
import os
import glob
```

# Preparação:

## Listas e elementos Impostantes:

```python
# Temporalidades:
ano_bissexto = ['2000', '2004', '2008', '2012', '2016', '2020']

anos = ['1997','1998','1999','2000','2001','2002','2003','2004','2005','2006',
        '2007','2008','2009','2010','2011','2012','2013','2014','2015','2016',
        '2017','2018','2019','2020','2021','2022']

meses = ['01','02','03','04','05','06','07','08','09','10','11','12']

# Estações:
estacoes = ['Vidigal','Urca','Rocinha','Tijuca','Santa Teresa','Copacabana',
            'Grajau','Ilha do Governador','Penha','Madureira','Iraja','Bangu',
            'Piedade','Tanque','Saude','Jardim Botanico','Barrinha',
            'Cidade de Deus','Riocentro','Guaratiba','Grajau Jacarepagua',
            'Santa Cruz','Grande Meier','Anchieta','Grota Funda','Campo Grande',
            'Sepetiba','Alto da Boa Vista','Av Brasil Mendanha','Recreio',
            'Laranjeiras','Sao Cristovao','Tijuca Muda']


#Diretórios:
diretorio = r'D:\Estudo\Oficina de Pesquisa\Dados Brutos'
diretorio_tratado = r'D:\Estudo\Oficina de Pesquisa\Dados Tratados'
diretorio_totais = r'D:\Estudo\Oficina de Pesquisa\Totais'

# Nome das Colunas - Para a importação:
colunas = ['Dia', 'Hora','HBV','15 min','01 h','04 h','24 h','96 h']

# Valor defaut para estações - Tem estação que não tem o valor, ai usa-se o valor default:
defaut = float(0)
```

# Correção Geral:

```python
qtd = 0 # Contagem das estações:
for estacao in estacoes:
    print(f'iniciando tratamendo da estação {estacao}')
    #Tratando as entradas:
    arquivo = trata_palavra(estacao)
    qtd += 1 #Essa contagem aqui serve só no final, pra uma informação visual se foi tudo certinho.
    
    # Criando a pasta da estacao caso ela não exista
    estacao_dir = os.path.join(diretorio_tratado, estacao)
    if not os.path.exists(estacao_dir):
        os.mkdir(estacao_dir)
        
    for ano in tqdm(anos):
        for mes in meses:
            try:
                # Lendo um arquivo:
                with open(diretorio+'\\'+estacao+'\\'+arquivo+'_'+ano+mes+'_Plv.txt', 'r') as arq_r:
                    linhas = arq_r.readlines()
                
                # Salvando em outro arquivo:
                with open(estacao_dir+'\\'+arquivo+'_'+ano+mes+'_Plv.txt', 'w') as arq_w:
                    for i, linha in enumerate(linhas):
                        if i>=5:
                            arq_w.write(linha)
            except:
                print(f'A estação {estacao} não possui o mes {mes} do ano {ano}')
    print(f'Tratamento da estação {estacao} concluido :)\n')
    
# Checagem visual:
checa_quantidade(qtd,len(estacoes)) # Para saber se todas as estações foram corrigidas:
```

# Manipulação dos dados:

## Tabelas com o Fevereiro de 31 dias:

```python
# Criando os DataFrames:
qtd = 0 # Contagem das estações:
for estacao in estacoes:
    # Listas - Valores Absolutos:
    soma_fev = []
    soma_fev31 = []
    soma_mar = []
    
    # Listas - Valores Nomalizados
    soma_fevN = []
    soma_fev31N = []
    soma_marN = []
    
    print(f'Iniciando construção do DataFrame da estação {estacao}...')
    #Tratando as entradas:
    arquivo = trata_palavra(estacao)
    qtd += 1 #Essa contagem aqui serve só no final, pra uma informação visual se foi tudo certinho.
    
    estacao_dir = os.path.join(diretorio_tratado, estacao)
    
    for ano in tqdm(anos):
        try:
            df_fev = pd.read_fwf(estacao_dir+'\\'+arquivo+'_'+ano+'02_Plv.txt', 
                                     colspecs = [(0,10),(12,20),(22,25),(28,34),(35,41),(42,48),(49,55),(56,62)],
                                     infer_datetime_format=True, enconding='iso-8850-1',
                                     names = colunas)
                
            df_mar = pd.read_fwf(estacao_dir+'\\'+arquivo+'_'+ano+'03_Plv.txt', 
                                 colspecs = [(0,10),(12,20),(22,25),(28,34),(35,41),(42,48),(49,55),(56,62)],
                                 infer_datetime_format=True, enconding='iso-8850-1',
                                 names = colunas)
            
            # Construção do Fevereiro de 31 Dias:
            mes_concat = pd.concat([df_fev, df_mar], ignore_index = False)
            
            # Checando os anos bissextos:
            if ano in ano_bissexto:
                mes_concat['Dia'] = pd.to_datetime(mes_concat['Dia'], dayfirst= True, infer_datetime_format=True)
                meses_id = mes_concat.set_index('Dia')
                fev31 = meses_id.loc[: ano+"-03-02"]
            else:
                mes_concat['Dia'] = pd.to_datetime(mes_concat['Dia'], dayfirst= True, infer_datetime_format=True)
                meses_id = mes_concat.set_index('Dia')
                fev31 = meses_id.loc[: ano+"-03-03"]
            
            fev31 = fev31.reset_index()
            
            # Fevereiro 31 Dias:
            fev31['15 min'] = pd.to_numeric(fev31['15 min'], errors='coerce')
            fev31['15 min'].fillna(fev31['15 min'].mean(), inplace=True)
            soma = fev31['15 min'].sum()
            norm = soma/31
            soma_fev31.append(soma)
            soma_fev31N.append(norm)
            
            # Março:
            df_mar['15 min'] = pd.to_numeric(df_mar['15 min'], errors='coerce')
            df_mar['15 min'].fillna(df_mar['15 min'].mean(), inplace=True)
            soma = df_mar['15 min'].sum()
            norm = soma/31
            soma_mar.append(soma)
            soma_marN.append(norm)
            
            # Fevereiro Natural:
            df_fev['15 min'] = pd.to_numeric(df_fev['15 min'], errors='coerce')
            df_fev['15 min'].fillna(df_fev['15 min'].mean(), inplace=True)
            soma = df_fev['15 min'].sum()
            soma_fev.append(soma)
            if ano in ano_bissexto:
                norm = soma/29
                soma_fevN.append(norm)
            else:
                norm = soma/28
                soma_fevN.append(norm)
            
            
            df = pd.DataFrame({'FevNatural':soma_fev,'Fev31':soma_fev31,'Março':soma_mar,
                               'FevNaturalN':soma_fevN,'Fev31N':soma_fev31N,'MarçoN':soma_marN})
            df_anos = pd.DataFrame({'Anos':anos})
            df = pd.concat([df_anos,df], axis=1)
            df.to_csv(diretorio_tratado+'\\'+arquivo+'_Final.csv', index = False)
        except:
            soma_fev.append(defaut)
            soma_fev31.append(defaut)
            soma_mar.append(defaut)
            soma_fevN.append(defaut)
            soma_fev31N.append(defaut)
            soma_marN.append(defaut)
            print(f'A estação {estacao}, no ano {ano}, não possui algum dos meses necessários')
    display(df)
    print(f'DataFrame da estação {estacao} construido.')
    
# Checagem visual:
checa_quantidade(qtd,len(estacoes))  # Para saber se foi contruida tabela de todas as estações:
```

## Gerando o total de Chuva anual por mes de cada Estação:

```python
# Lista com os nomes dos meses:
meses_name = ['janeiro', 'fevereiro', 'março', 'abril', 'maio', 'junho',
              'julho', 'agosto', 'setembro', 'outubro', 'novembro', 'dezembro']

# Criando os DataFrames:
qtd = 0 # Contagem das estações:
for estacao in tqdm(estacoes): 
    #Criando o Dicionário:
    meses_dict = {mes: [] for mes in meses_name}

    # Criação do dicionário que associa o número do mês ao nome do mês:
    meses_dict_nomes = {mes: nome for mes, nome in zip(meses, meses_name)}
    
    print(f'Iniciando construção do DataFrame da estação {estacao}...')
    arquivo = trata_palavra(estacao) # Tratando as entradas
    qtd += 1 # Essa contagem aqui serve só no final, pra uma informação visual se foi tudo certinho.
    estacao_dir = os.path.join(diretorio_tratado, estacao)
    
    for ano in anos:
        for mes in meses:
            nome_mes = meses_dict_nomes[mes] # Obtem o nome do mês a partir do número do mês
            try:
                df_mes = pd.read_fwf(estacao_dir+'\\'+arquivo+'_'+ano+mes+'_Plv.txt', 
                                         colspecs = [(0,10),(12,20),(22,25),(28,34),(35,41),(42,48),(49,55),(56,62)],
                                         infer_datetime_format=True, enconding='iso-8850-1',
                                         names = colunas)
                
                # somatório do respectivo Mes:
                df_mes['15 min'] = pd.to_numeric(df_mes['15 min'], errors='coerce')
                df_mes['15 min'].fillna(df_mes['15 min'].mean(), inplace=True)
                soma = df_mes['15 min'].sum()
                meses_dict[nome_mes].append(soma)
            except:
                meses_dict[nome_mes].append(defaut)
                #print(f'A estação {estacao}, no ano {ano}, não possui algum dos meses necessários')
    print(f'DataFrame da estação {estacao} finalizado e exportado!')        
    df = pd.DataFrame(meses_dict)
    df_anos = pd.DataFrame({'Anos':anos})
    df = pd.concat([df_anos,df], axis=1)
    df.to_csv(diretorio_totais+'\\'+arquivo+'_Totais.csv', index = False)
display(df)
```
